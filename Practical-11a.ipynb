## Word Vector Extraction - Extract the word vector representation of the word using word2vec model

```python
import spacy

#Load the medium-sized English model
nlp=spacy.load("en_core_web_md")
```

```python
#Input word
word=input("Enter a word: ")
```

```python
# Process the word with spacy
token=nlp(word)[0]
```

```python
#Extract the vector
vector=token.vector
```

```python
print(f"Shape of the vector: {vector.shape}")
print()
print(f"Vector for the word '{word}':\n{vector}")
```
